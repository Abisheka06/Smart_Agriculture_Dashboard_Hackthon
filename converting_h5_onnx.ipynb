```python
import tensorflow as tf
import tf2onnx
from tensorflow.keras.models import load_model
from tensorflow.keras import Model
from google.colab import files

# Upload the model
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Load and wrap model
seq_model = load_model(filename)

# Wrap in a Model if it's Sequential
inputs = tf.keras.Input(shape=seq_model.input_shape[1:], name="input")
outputs = seq_model(inputs)
model = Model(inputs=inputs, outputs=outputs)

# Convert to ONNX
spec = (tf.TensorSpec(model.inputs[0].shape, tf.float32, name="input"),)
onnx_model, _ = tf2onnx.convert.from_keras(model, input_signature=spec)

# Save to ONNX file
onnx_filename = filename.replace(".h5", ".onnx")
with open(onnx_filename, "wb") as f:
    f.write(onnx_model.SerializeToString())

# Download
files.download(onnx_filename)
```
